```python
from ultralytics import YOLOv10
import supervision as sv
import cv2


model = YOLOv10('best.pt')
```

```python
def Draw(Image_path):
    image = cv2.imread(Image_path)
    results = model(image)[0]
    detections = sv.Detections.from_ultralytics(results)

    bounding_box_annotator = sv.BoxAnnotator()
    label_annotator = sv.LabelAnnotator()

    annotated_image = bounding_box_annotator.annotate(
        scene=image, detections=detections)
    annotated_image = label_annotator.annotate(
        scene=annotated_image, detections=detections)

    # Affichage de l'image annotée
    cv2.imshow("Annotated Image", annotated_image)
    cv2.waitKey(0)  # Attendre jusqu'à ce que l'utilisateur appuie sur une touche
```

```python
#results = Run_Image("Images/Image3.jpg")
Draw("Images/Image1.jpg")
```

```python
def Draw_Video(video_path):
    cap = cv2.VideoCapture(video_path)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame)[0]
        detections = sv.Detections.from_ultralytics(results)

        bounding_box_annotator = sv.BoxAnnotator()
        label_annotator = sv.LabelAnnotator()

        annotated_frame = bounding_box_annotator.annotate(
            scene=frame, detections=detections)
        annotated_frame = label_annotator.annotate(
            scene=annotated_frame, detections=detections)

        cv2.imshow('Annotated Video', annotated_frame)

        # Quitter si l'utilisateur appuie sur 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```

```python
Draw_Video("Images/sstt.mp4")
```

```python
def Draw_Live():
    cap = cv2.VideoCapture(0)  # Utiliser 0 pour la caméra par défaut

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame)[0]
        detections = sv.Detections.from_ultralytics(results)

        bounding_box_annotator = sv.BoxAnnotator()
        label_annotator = sv.LabelAnnotator()

        annotated_frame = bounding_box_annotator.annotate(
            scene=frame, detections=detections)
        annotated_frame = label_annotator.annotate(
            scene=annotated_frame, detections=detections)

        cv2.imshow('Live Annotated Video', annotated_frame)

        # Quitter si l'utilisateur appuie sur 'q'
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()
```
